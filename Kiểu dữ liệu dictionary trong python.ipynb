```python
#Ví dụ
SinhVien = {
    "HoVaTen" : "Nguyễn Văn A",
    "MaLop" : "DH07",
    "DiemTrungBinh" : 8.5
}
print(SinhVien)
print(SinhVien["HoVaTen"])
```

```python
#Sử dụng get() để lấy giá trị
print(SinhVien.get("MaLop"))
```

```python
#Cập nhật, thay đổi giá trị => dict["keyn"] = "value mới"
SinhVien["MaLop"] = "DH02"
print(SinhVien)
# => dict.update({"key1":"value mới 1", "key2":"value mới 2",...})
SinhVien.update({"MaLop":"DH03", "DiemTrungBinh": 8.6})
print(SinhVien)
```

```python
#Thêm các mục : Gần giống với cách cập nhật, nhưng cặp key - value mới
SinhVien["NamHoc"] = 2025
print(SinhVien)

SinhVien.update({"NoiSinh":"Nam Định"})
print(SinhVien)
```

```python
#Loại bỏ các mục 
SinhVien.pop("NoiSinh") #xóa cặp key - value được chỉ định bằng việc nhập keyword vào hàm
print(SinhVien)

SinhVien.popitem() #tự động xóa cặp key - value cuối cùng
print(SinhVien)

del SinhVien["HoVaTen"]
print(SinhVien)
```

```python
#Sử dụng vòng lặp 
SinhVien = {
    "HoVaTen" : "Nguyễn Văn A",
    "MaLop" : "DH07",
    "DiemTrungBinh" : 8.5
}
# In tất cả các từ khóa trong từ điển, từng cái một
for x in SinhVien:
    print(x)
    
# Duyệt các giá trị
for x in SinhVien.values():
    print(x)
    
# Duyệt các khóa
for x in SinhVien.keys():
    print(x)

# Duyệt các cặp khóa - giá trị     
for x, y in SinhVien.items():
    print(x,y)
```

```python
"""
Bài tập: Xây dựng một từ điển có các chức năng như sau ( người dùng,lựa chọn, chức năng thông qua menu)
1. Thêm một từ vựng mới kèm nghĩa của từ vựng vào từ điển
2. Tra cứu ý nghĩa của một từ vựng
3. Cập nhật ý nghĩa cho một từ vựng
4. Cho phép người dùng xóa đi một từ vựng trong từ vựng
5. Cho phép người dùng xóa đi toàn bộ từ vựng
6. Cho phép người dùng in ra toàn bộ từ vựng
7. Cho phép người dùng in ra toàn bộ từng điển theo cấu trúc: "TỪ VỰNG" : "Ý NGHĨA"
8. Kết thúc chương trình
"""
```

```python
#Bài giải
tudien = {} #Khai báo từ điển

while(True):
    print("Vui lòng lựa chọn chức năng (bằng số) ")
    print("""
    1. Thêm một từ vựng mới kèm nghĩa của từ vựng vào từ điển\n
    2. Tra cứu ý nghĩa của một từ vựng\n
    3. Cập nhật ý nghĩa cho một từ vựng\n
    4. Cho phép người dùng xóa đi một từ vựng trong từ vựng\n
    5. Cho phép người dùng xóa đi toàn bộ từ vựng\n
    6. Cho phép người dùng in ra toàn bộ từ vựng\n
    7. Cho phép người dùng in ra toàn bộ từng điển theo cấu trúc: "TỪ VỰNG" : "Ý NGHĨA"\n
    8. Kết thúc chương trình)\n
    """)
    luachon = int(input("Nhập vào lựa chọn của bạn: "))
    if luachon == 1 or luachon == 3:
        tuvung = input("Nhập vào từ vựng: ")
        ynghia = input("Nhập vào ý nghĩa: ")
        tudien[tuvung] = ynghia
        print("Đã cập nhật dữ liệu")
    
    elif luachon == 2:
        tuvung = input("Nhập vào từ vựng cần tra: ")
        print("Ý nghĩa: ", tudien[tuvung])
    
    elif luachon == 4:
        tuvung = input("Nhập vào từ vựng cần xóa: ")
        tudien.pop(tuvung)
        print("Đã xóa dữ liệu")
        
    elif luachon == 5: 
        tudien.clear()
        print("Đã xóa toàn bộ dữ liệu")
    
    elif luachon == 6: 
        print("Danh sách các từ vựng có trong từ điển: ")
        for x in tudien.keys(): 
            print(x)
        
    elif luachon == 7:
        print("Danh sách các từ vựng có trong từ điển: ")
        for x, y in tudien.items():
            print(x, ":", y)
    
     elif luachon == 8:
        break;
    
    else:
        print("Lựa chọn không hợp lệ")
```
