- Set là một trong 4 kiểu dữ liệu được tích hợp sẵn trong python
- Set là một tập hợp không có thứ tự, giá trị không được trùng nhau và không thể thay đổi và không được lập chỉ mục
- Lưu ý: Mục set là không thể thay đổi, nhưng chúng ta có thể xóa và thêm các mục mới
- Sử dụng cặp ngoặc { }

Ví dụ: 

```python
#Khởi tạo set: Set = set()
monhoc = {"Toán", "Lý", "Hóa"}
print(monhoc)
```

```python
#Duyệt phần tử trong set
for x in monhoc:
    print(x)
```

```python
#Thêm mới các phần tử trong set
#Thêm một phần tử: sử dụng add()
monhoc.add("Lịch Sử")
#Thêm một phần tử khác hoặc một list khác: sử dụng update()
hocthem = {"Anh Văn", "Đàn Piano"}
monhoc.update(hocthem)
print(monhoc)
```

```python
#Thêm một list vào một set
hocphudao = ["Võ thuật", "Múa", "Võ thuật"]
monhoc.update(hocphudao)
print(monhoc)
```

```python
#Xóa phần tử
monhoc.remove("Sinh học")
```

```python
monhoc.remove("Võ thuật")
monhoc.discard("Toán")
print(monhoc)
```

```python
- Xóa phần tử:
    + Sử dụng phương thức remove(): nếu phần tử không tồn tại sẽ phát sinh lỗi
    + Sử dụng phương thức discard(): nếu phần tử không tồn tại sẽ không phát sinh lỗi 
    + Loại bỏ phần tử đầu tiên bằng phương thức pop()
    + Làm rông phần tử bằng phương thức clear()
    + Xóa bỏ tập hợp bằng từ khóa del 
```

```python
monhoc.pop()
print(monhoc)
```

```python
monhoc.clear()
print(monhoc)
```

```python
del monhoc
print(monhoc)
```
