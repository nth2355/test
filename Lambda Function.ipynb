- Một lambda là một hàm ẩn đanh nhỏ.
- Một hàm lambda có thể nhận bất kì số lượng số nào, nhưng chỉ có thể có một biểu thức


# Cú pháp

lambda argument: expression

```python
#Ví dụ: 
kiemtrasochan = lambda a : (a%2==0)
print(kiemtrasochan(5))
print(kiemtrasochan(4))
```

```python
tinhtong = lambda a, b : a+b
print(tinhtong(5, 10))
print(tinhtong(-5, 2))
```

```python
#tương tự với phương pháp định nghĩa hàm
def tinhtong (a,b):
    return a+b
print(tinhtong(5,10))
```

# Ví dụ về sử dụng lambda function bên trong các funtion

```python
#hàm mũ
def hammuhai(a):
    return a**2
def hammuba(a):
    return a**3
#Dùng lambda
def hammu(n):
    return lambda x : x**n
hammu2 = hammu(2)
hammu3 = hammu(3)

print(hammu2(3))
print(hammu2(4))
print(hammu3(5))
```
