# Xử lý ngoại lệ trong Python


```python
try:
# Đoạn code dự đoán có lỗi
except:
# Hành động khi lỗi xảy ra
else:
# Thực thi đoạn này nếu như mã không có lỗi
finally:
# Cho phép bạn thực thi mã, bất kể kết quả của các khối try có bị lỗi hay không
```

Ví dụ:

```python
a = int(input("Nhập vào số nguyên a: "))
print(str(a) + " + 5 = "+ str(a+5))
```

Cũng với ví dụ trên, ta hãy nhập dữ liệu dưới dạng chữ

```python
a = int(input("Nhập vào số nguyên a: "))
print(str(a) + " + 5 = "+ str(a+5))
```

Như quan sát thấy, khi ta nhập một dữ liệu khác dưới dạng khác với kiểu dữ liệu "int" được ép ban đầu cho dữ liệu, chương trình sẽ ra lỗi. Những lỗi này không phải do dev hay lập trình viên cố ý, mà chúng là những ngoại lệ ở thực tế mà khi lập trình, ta chưa lường trước được.
Thế nên, khi lập trình, ta phải lường trước những "lỗi" này và thực hiện các thao tác khi xảy ra các lỗi ấy.
Và 'try - catch except' sẽ giúp chúng ta làm điều đó

```python
try:
    a = int(input("Nhập vào số nguyên a: "))
    print(str(a) + " + 5 = "+ str(a+5))
except:
    print("Nhập dữ liệu không chính xác!")
#else:
# Thực thi đoạn này nếu như mã không có lỗi
#finally:
# Cho phép bạn thực thi mã, bất kể kết quả của các khối try có bị lỗi hay không
```

Kết quả cho thấy, khi chúng ta sử dụng try-catch except, đoạn mã sẽ không báo lỗi khi ta nhập một kiểu dữ liệu sai với yêu cầu, mà chương trình sẽ thông báo đoạn mã trong khối 'except' -> một yêu cầu khi đoạn mã nhận dữ liệu ngoại lệ

Hoặc chúng ta có thể phát hiện lỗi tự động bằng cách sau:


```python
try:
    a = int(input("Nhập vào số nguyên a: "))
    print(str(a) + " + 5 = "+ str(a+5))
except Exception as e:
    print(e)
```

```python
try:
    a = int(input("Nhập vào số nguyên a: "))
    print(str(a) + " + 5 = "+ str(a+5))
except Exception as e:
    print(e)
else:
    print("Không có lỗi xảy ra")
finally:
    print("Kết thúc chương trình")
```

Đối với hai khối mã 'else' và 'finally':
    - Khối 'else' sẽ được thực thi khi khối 'try' không phát hiện lỗi và khối except không chạy
    - Khối 'finally' luôn chạy dù khối mã ở khối 'try' lỗi hay không lỗi
    - Xem ví dụ trên để hiểu hơn
