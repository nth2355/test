- Nếu chúng ta không biết có bao nhiêu đối số sẽ được truyển, chúng ta thêm một dấu * trước tên tham số trong định nghĩa hàm. Hàm sẽ nhận được nhiều đối số và có thể truy cập các mục tương ứng:


```python
def thoikhoabieu(*monhoc):
    print("Môn 1: " + monhoc[0])
    print("Môn 2: " + monhoc[1])
```

```python
thoikhoabieu("Toán", "Lý", "Hóa", "Sinh", "Sử", "Địa")
```

```python
def tinhtong(*giatri):
    sum = 0
    for x in giatri:
        sum+=x
    print(sum)
```

```python
tinhtong(1, 2)
tinhtong(1, 4, 7, 8, 9, 5)
```

- Sử dụng từ khóa cho nhiều đối số, thêm dấu ** vào trước tên tham số

```python
def xinchao(**hovaten):
    print("Xin chào! " + hovaten["ten"])
```

```python
xinchao(ten="Hải", chulot="Thanh", ho="Nguyễn")
```

- Sử dụng từ khóa "return" để trả về giá trị

```python
def tinhtich(*giatri):
    tich = 1
    for x in giatri:
        tich*=x
    return tich
```

```python
tich1 = tinhtich(1, 4, 6)
tich2 = tinhtich(7, 4, 6)
tong = tich1 + tich2
print(tong)
```

# Bài tập

- Bài tập: Tìm USCLN của 2 số a, b

```python
#Xây dựng hàm gcd(a,b) => Trả về USCLN
#Ví dụ: 1, 13 => gcd(1,13) = 1
#Ví dụ: 35, 77 => gcd(35,77) = 7
# Thuật toán sử dụng: Trong 2 đối số a và b, lấy số lớn hơn trừ cho số bé hơn, thao tác nỳ sẽ lặp đi lặp lại cho đến khi hai số bằng nhau
# Ví dụ: gcd(35, 77): 
    #35,42
    #35,7
    #28,7
    #21,7
    #14,7
    #7,7
def gcd(a, b):
    while a!=b:
        if a > b:
            a-=b
        else:
            b-=a
    return a
    
```

```python
print(gcd(55,100))
print(gcd(11,121))
```

- Bài tập: Nhập vào một dãy n số từ bàn phím, sau đó tính tổng.

```python
#Yêu cầu
#Xây dựng các hàm:
#nhạp(n, list_number)
#tinhtong(list_number)

#Khai báo biến
list_number = []
n = 1
while(True):
    try: 
        n = int(input("Nhập vào số lượng phần tử: "))
    except: 
        print("Vui lòng nhập n >= 1")
    if n>=1:
        break


#Hàm nhập()
def nhap(n, list_number):
    for i in range(n):
        list_number.append(int(input("Nhập vào giá trị thứ: " + str(i) + ":")))

#Hàm tính tổng
def tinhtong(list_number):
    tong = 0
    for x in list_number:
        tong+=x
    return tong

nhap(n, list_number)
print("Tổng = " + str(tinhtong(list_number)))
```
