```python
#Tạo ra list có dữ liệu
colors = ["red", "green", "orange"]
print(colors)
```

```python
#List có thứ tự vị trí các phần tử được đánh dáu từ trái sang phải
studentList = ["An", "Binh", "Ngân", "Vy"]
print(studentList[2])

#Muốn in ra màn hình một phần tử trong List, t thực hiện cú pháp: print(List[số thứ tự phần tử đếm từ số 0])
print(studentList[0])
```

```python
#Cách in ra cả List
print(studentList)
print(studentList[:])
```

```python
#studentList[x:y] => Lấy ra các phần tử có số thứ tự nguyên trong nửa khoảng [x,y)
print(studentList[1:2])
print(studentList[0:2])
print(studentList[1:4])
```

```python
#Thêm phần tử vào cuối list
studentList.append("Thiên")
print(studentList)
# => Để thêm một phần tử vào cuối danh sách, ta dùng cú pháp list.append(phần tử)
studentList[len(studentList):] = ["Thành"]
print(studentList)
```

```python
#Chèn phần tử vào list
studentList.insert(2, "Ngọc")
print(studentList)
# => Để chèn một phần tử khác vào vị trí bất kì, ta dùng cú pháp list.insert(vị trí, phần tử)
```

```python
#Số lượng phần tử có trong list: => len(list)
print(len(studentList))
```

```python
#Đếm số lượng phần tử thỏa điều kiện => list.count(phần tử)
print("Đếm Ngọc: ", studentList.count("Ngọc"))
```

```python
#Xóa phần tử ra khỏi list bằng phần tử => Cú pháp: list.remove(phần tử)
#Nếu có nhiều hơn một phần tử giống nhau sẽ xóa theo thú tự từ trái sang phải
studentList.remove("Ngân")
print(studentList)
```

```python
#kiểm tra phần tử có trong list hay không: If x in list
if "Ngọc" in studentList:
    studentList.remove("Ngọc")
    print(studentList)
# => ưu điểm: hàm sẽ không bị lỗi nếu không có phần tử trong List 

#Xóa phần tử ra khỏi list bằng vị trí => List.pop(vị trí)
studentList.pop(0)
print(studentList)
```

```python
#Đảo ngược list => list.reverse()
studentList.reverse()
print(studentList)
```

```python
#Sắp xếp lại => list.sort()
studentList.sort()
print(studentList)
```

```python
numbers = [7, 5, 1, 9, 0, 5, 7]
numbers.sort()
print(numbers)
```

```python
#sắp xếp ngược, sắp xếp giảm dần => list.sort(reverse=True)
studentList.sort(reverse=True)
print(studentList)
numbers.sort(reverse=True)
print(numbers)
```

```python
#xóa sạch dữ liệu trong list => Cú pháp: list.clear()
studentList.clear()
print(studentList)
```

```python
Tóm lại:
    - Tạo list rỗng:
        List = []
        List = list()
    - Cách in ra cả một list:
        print(List)
        print(List[:])
    - Cách in ra các phần tử được chọn:
        print(List[x:y]) với x, y là thứ tự của phần tử tương ứng
        => in ra các giá trị có thứ tự i sao cho x =< i <y
    - Cách thêm phần tử vào cuối List:
        List.append(x)
        List[len(List):] = [x] với x là GIÁ TRỊ được thêm vào
    - Cách chèn phần tử vào vị trị bất kì:
        List.insert(x, y) với:
            x là thứ tự chèn
            y là GIÁ TRỊ được chèn
    - Cách đếm số lượng phần tử trong List:
        print(len(List))
    - Cách đếm số phần tử thỏa điều kiện:
        print(List.count(x)) với x là GIÁ TRỊ cần đếm
    - Cách xóa phần tử trong List:
        + Xóa bằng GIÁ TRỊ:
            List.remove(x) với x là GIÁ TRỊ cần xóa
        + Xóa bằng số thứ tự:
            List.pop(y) với y là thứ tự đếm từ 0 của phần tử cần xóa 
        Có thể kết hợp với cú pháp điều kiện để tránh lỗi hàm khi phần tử cần xóa không có trong list
    - Cách đảo ngược hàm:
        List.reverse()
    - Cách sắp xếp lại List (nếu là chuỗi sẽ sắp xếp theo alphabet, nếu là số sẽ sắp xếp từ bé đến lớn):
        List.sort()
    - Cách sắp xếp ngược (chuỗi kí tự) , sắp xếp giảm dần (chuỗi số):
        List.sort(reverse=True)
    - Cách xóa sạch dữ liệu trong List:
        List.clear()
```
